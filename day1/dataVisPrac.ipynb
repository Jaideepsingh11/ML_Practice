```python
xaxis = np.arange(0, 10, 1)

# create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# -------- LINE WIDTH EXAMPLES --------
ax.plot(xaxis, xaxis + 1, color="red", linewidth=0.25)
ax.plot(xaxis, xaxis + 2, color="red", linewidth=0.50)
ax.plot(xaxis, xaxis + 3, color="red", linewidth=1.00)
ax.plot(xaxis, xaxis + 4, color="red", linewidth=2.00)

# -------- LINE STYLE EXAMPLES --------
ax.plot(xaxis, xaxis + 5, color="green", linewidth=3, linestyle='-')   # solid
ax.plot(xaxis, xaxis + 6, color="green", linewidth=3, linestyle='--')  # dashed
ax.plot(xaxis, xaxis + 7, color="green", linewidth=3, linestyle=':')   # dotted

# -------- CUSTOM DASH STYLE --------
line, = ax.plot(xaxis, xaxis + 8, color="black", linewidth=1.5)
line.set_dashes([5, 5, 15, 5])  
# format: [line length, gap, line length, gap]

# -------- MARKER EXAMPLES --------
ax.plot(xaxis, xaxis + 9, color="blue", linewidth=3, linestyle='--', marker='+')
ax.plot(xaxis, xaxis + 10, color="blue", linewidth=3, linestyle='--', marker='o')
ax.plot(xaxis, xaxis + 11, color="blue", linewidth=3, linestyle='--', marker='s')
ax.plot(xaxis, xaxis + 12, color="blue", linewidth=3, linestyle='--', marker='1')

# -------- MARKER SIZE + COLORS --------
ax.plot(xaxis, xaxis + 13, color="purple", linewidth=1, linestyle='-', marker='o', markersize=2)
ax.plot(xaxis, xaxis + 14, color="purple", linewidth=1, linestyle='-', marker='o', markersize=4)
ax.plot(xaxis, xaxis + 15, color="purple", linewidth=1, linestyle='-', marker='o', markersize=8, markerfacecolor="red")

# -------- FULL CUSTOM MARKER --------
ax.plot(
    xaxis,
    xaxis + 16,
    color="purple",
    linewidth=1,
    linestyle='-',
    marker='s',
    markersize=8,
    markerfacecolor="yellow",
    markeredgewidth=3,
    markeredgecolor="green"
)

# show plot
plt.show()
```
